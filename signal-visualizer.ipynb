```python
import numpy as np
import matplotlib.pyplot as plt

def sinwavegen(freq, amp, duration, sample_rate=1000):
    time = np.linspace(0, duration, int(duration * sample_rate))
    voltage = amp * np.sin(2 * np.pi * freq * time)
    return time, voltage

def plot_signal(time, voltage):
    plt.plot(time, voltage)
    plt.title("AC Sine Wave Sim")
    plt.xlabel("Time (s)")
    plt.ylabel("Voltage (V)")
    plt.grid(True)
    plt.savefig("signal.png")
    plt.show()

f = int(input("Enter the frequency:"))
a = int(input("Enter the amplitude:"))
d = int(input("Enter the duration:"))

t, v = sinwavegen(f, a, d)
plot_signal(t, v)
```
