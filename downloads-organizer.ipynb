```python
import os
import shutil

target_directory = os.path.expanduser("~/Downloads")

extensions = {'.pdf':'Documents' , '.docx':'Documents' , '.jpg':'Images' , '.jpeg':'Images' , '.png':'Images' , '.exe':'Apps'}

def organize (directory_path):
    
    for file_name in os.listdir(directory_path):
        file_path = os.path.join(directory_path, file_name)

        if os.path.isdir(file_path):
            continue

        name, extension = os.path.splitext(file_name)

        if extension in extensions:
            folder_name = extensions[extension]
            destination_folder = os.path.join(directory_path, folder_name)

            if not os.path.exists(destination_folder):
                os.makedirs(destination_folder)

            destination_path = os.path.join(destination_folder, file_name)
            shutil.move(file_path, destination_path)

            print('Completed')

organize(target_directory)
```
