```python
import numpy as np
import matplotlib.pyplot as plt

def charging_sim (v, r, c):
    time_const = r*c
    time = np.linspace (0, 5*time_const, 1000)
    v_cap = v * (1 - (np.exp(-time/time_const)))
    return time, v_cap

def plot_graph (time, v_cap):
    plt.plot (time, v_cap)
    plt.title ("RC charging graph")
    plt.xlabel ("Time (s)")
    plt.ylabel ("Voltage (V)")
    plt.savefig ("RC_circuit.png")
    plt.show()

voltage = int(input("Enter the voltage: "))
resistance = int(input("Enter the resistance: "))
capacitance = float(input("Enter the capacitance: "))

t, vc = charging_sim(voltage, resistance, capacitance)
plot_graph(t, vc)
```
